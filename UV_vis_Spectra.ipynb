```python
#Use this cell if you have multiple .csv files and absorption spectra to plot
#Pay attention to skip_header

aliquots = 1
absorb = [0] * aliquots

#Include the file path to ur .csv file. Example below:
uvvis1 = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/25' Amisha Jain/Summer 2022/AJ07.2 UV-Vis Data/Modified Data (.CSV)/Aliquot_1_0C.csv"
absorb[0]= np.genfromtxt(uvvis1, skip_header = 1, delimiter = ',', dtype =np.float)

uvvis2 = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/"
absorb[1]= np.genfromtxt(uvvis2, skip_header = 1, delimiter = ',', dtype =np.float)

uvvis3 = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/"
absorb[2]= np.genfromtxt(uvvis3, skip_header = 1, delimiter = ',', dtype =np.float)

uvvis4 = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/"
absorb[3]= np.genfromtxt(uvvis4, skip_header = 1, delimiter = ',', dtype =np.float)

uvvis5 = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/"
absorb[4]= np.genfromtxt(uvvis5, skip_header = 1, delimiter = ',', dtype =np.float)

#Use newAbsorb if you have baseline file(s) to subtract
numFiles = 1
newAbsorb = [0] * numFiles

baselines = "/content/drive/MyDrive/Olshansky Lab Docs/Lab Notebooks/"
newAbsorb[0] = np.genfromtxt(baselines, skip_header = 1, delimiter = ',', dtype =np.float)
```

```python
#To plot the same absorption spectrum column from each .csv aliquot file

%matplotlib inline
legendlist = []
for i in range(0, aliquots):
    plt.plot(absorb[i][:,0],absorb[i][:,1])
    plt.xlim(300,380) #you can change these limits to get a better look
    plt.ylim(0,1.7)
    plt.xlabel('Wavelength / nm', fontsize=15)
    plt.ylabel('Absorbance / A.u.',fontsize=15)

#Adjust amount of labels depending on number of aliquots
legendlist.append("1")
legendlist.append("2")
legendlist.append("3")
legendlist.append("4")
legendlist.append("5")
plt.legend(legendlist)
plt.show()
```

```python
#To plot all the columns of absorption spectra in a single .csv aliquot file

#exclude wavelength column
columns = 5

%matplotlib inline
legendlist = []
for i in range(0, columns):
    plt.plot(absorb[0][:,0],absorb[0][:,i+1]) #put in the file you want (ex. absorb[0])
    plt.xlim(200,500) #you can change these limits to get a better look
    plt.ylim(0,3.5)
    plt.xlabel('Wavelength / nm', fontsize=15)
    plt.ylabel('Absorbance / A.u.',fontsize=15)

#Adjust amount of labels depending on number of columns
legendlist.append("1")
legendlist.append("2")
legendlist.append("3")
legendlist.append("4")
legendlist.append("5")
plt.legend(legendlist)
plt.show()
```
